## Segmentation for Cars Dataset

```python
# Data Preprocessing
# - Select relevant numeric features for segmentation
# - Handle missing data (if any)
# - Standardize the features (important for K-means)
selected_features_cars = ['Battery Capacity(kWh)', 'Acceleration(sec)', 'TopSpeed(km/h)', 'Range(km)', 'Max Power(kW)', 'Max Torque(Nm)', 'No. of Seats', 'Charging T(h)', 'Price(Lh)']
cars_data = cars_data[selected_features_cars].dropna()
```

```python
# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(cars_data)
```

```python
# Prepare the data with the selected features
X = cars_data[selected_features_cars]
```

```python
# Initialize a list to store the explained variances
explained_variances = []
```

```python
# Determine the optimal number of clusters (K)
for k in range(1, 11):  # Trying K from 1 to 10
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    explained_variances.append(kmeans.inertia_)  # Inertia measures how far the points within a cluster are
```

```python
# Plot the explained variances
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), explained_variances, marker='o', linestyle='--')
plt.title('Elbow Method for K-means Clustering (Cars Dataset)')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Explained Variance')
plt.grid(True)
plt.show()
```

```python
# Optimal value of k is obtained as 4
k = 4
```

```python
# Apply K-means clustering
kmeans = KMeans(n_clusters=k, random_state=42)
clusters = kmeans.fit_predict(scaled_data)
```

```python
# Add cluster labels to the original dataset
cars_data['Cluster'] = clusters
```

```python
# Principal Component Analysis (PCA) for visualization
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)
cars_data['PCA1'] = pca_result[:, 0]
cars_data['PCA2'] = pca_result[:, 1]
```

```python
# Visualize the clusters (2D PCA plot)
plt.figure(figsize=(10, 6))
plt.scatter(cars_data['PCA1'], cars_data['PCA2'], c=cars_data['Cluster'], cmap='viridis')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.title('Customer Segmentation (K-means)')
plt.colorbar()
plt.show()
```

```python
# Analyze the characteristics of each cluster
cluster_means = cars_data.groupby('Cluster').mean()
print(cluster_means)
```

Four clusters can be identified from the data, with the above mentioned average characteristics.

## Segmentation for Two-Wheeler Dataset

```python
# Select relevant numeric features for segmentation
numeric_features = ['Price', 'Range', 'Top_speed']
```

```python
two_wheelers_data = two_wheelers_data[numeric_features].dropna()
```

```python
# Create a pipeline to preprocess numeric features and perform clustering
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())  # Standardize numeric features
])
```

```python
# Define a column transformer to apply transformations to specific columns
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features)
    ])
```

```python
# Prepare the data with only the numeric features
X = two_wheelers_data[numeric_features]
```

```python
# Initialize a list to store the explained variances
explained_variances = []
```

```python
# Determine the optimal number of clusters (K)
for k in range(1, 11):  # Trying K from 1 to 10
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    explained_variances.append(kmeans.inertia_)  # Inertia measures how far the points within a cluster are
```

```python
# Plot the explained variances
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), explained_variances, marker='o', linestyle='--')
plt.title('Elbow Method for K-means Clustering (Two-Wheeler Dataset)')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Explained Variance')
plt.grid(True)
plt.show()
```

```python
# Optimal value of k is obtained as 4
k = 4
```

```python
# Apply K-means clustering within the pipeline
kmeans = Pipeline(steps=[('preprocessor', preprocessor),
                          ('model', KMeans(n_clusters=k, random_state=42))])
```

```python
# Fit the model and add cluster labels to the original dataset
two_wheelers_data['Cluster'] = kmeans.fit_predict(two_wheelers_data[numeric_features])
```

```python
# Analyze the characteristics of each cluster
cluster_means = two_wheelers_data.groupby('Cluster').mean()
print(cluster_means)
```

Four clusters/segements can be identified with the above mentioned average price, range and top speed

## Conclusion

### Cars Dataset

* Battery Capacity, acceleration, top speed, range, max power, max torque, no. of seats, charging, and price were selected as features.
* Considering these attributes, four clusters can be identified, with mean prices of 154.5L, 18L, 100L and 9L.
* This can be demarcate the market into budget, mid-level, high-level and luxury segments.
* Electric cars belonging to the 'luxury' segment (average price of 154L INR) have the highest battery capacity, longest range, quickest acceleration and top speed.
* As we go further down the segment line-up, these figures keep reducing, reaching the minimum for the budget segment (average price of 9L INR). Budget vehicles have an average acceleration of 8 seconds, while luxury vehicles take only 4.4 s.

### Two-Wheeler Dataset

* Based on price, range and top speed, four clusters can be identified, with average prices of 0.65L, 1.27L, 3.8L and 1.43L.
* These can be identified as belonging to the budget, mid-range, high-end and ultra high-end/high performance segments.
* Once again, a higher price gives longer range and greater speeds. The range for budget electric two-wheelers is 81km, while it is 307km for the highest-end products.
* However, we also note that lithium ion batteries are mainly used in electric two-wheelers. Alloy wheels also seem to be the standard.

Different customer bases can be identified based on the clusters obtained in order to better segment the car and two-wheeler EV market. We also note that while segments can be created using quantitative methods as has been done above, qualitative aspects (such as ride comfort, build quality, design, connectivity and technological features) must also be considered for product placement strategies and branding.
